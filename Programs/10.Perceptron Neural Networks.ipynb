Percepton Neural Network for AND gate using stochiastic gradient descent.

```python
import numpy as np
import matplotlib.pyplot as plt
x1 = np.array([0,0,1,1])
x2 = np.array([0,1,0,1])
yt = np.array([0,0,0,1])
yin = np.zeros(4)
ya = np.zeros(4, dtype=int)
w0 = 0
w1 = 1
w2 = 1
eta = 0.2
for j in range(0,10):
    for i in range(0,4):
        yin[i] = w0 + w1 * x1[i] + w2 * x2[i]
        if yin[i] > 0:
            ya[i] = 1
        else:
            ya[i] = 0
        if ya[i] != yt[i]:
            w0 = w0 + eta * (yt[i] - ya[i])
            w1 = w1 + eta * (yt[i] - ya[i]) * x1[i]
            w2 = w2 + eta * (yt[i] - ya[i]) * x2[i]
print("Expected output : ",yt)
print("Calculated output: ",ya)
y = (-w1/w2)*x1-(w0/w2)
plt.scatter(x1, x2,color = 'g')
plt.plot(x1,y,color='red')
plt.suptitle('Plot')
plt.show()
```
