PCA example on data from 2D to 2D reduction.

```python
import numpy as np
import matplotlib.pyplot as plt
from numpy import linalg as LA
data = np.array([[2, 1], [3, 4], [5, 0], [7, 6], [9, 2]])
X = data[:,0]
Y = data[:,1]
mean1 = np.mean(data, axis = 0)
data = data - mean1
cov = (np.dot(np.transpose(data), data)) / 4
w , v = LA.eig(cov)
data_proj = np.dot(data, v)
print("After dimension reduction feature matrix:")
print(data_proj)
new_x = data_proj[:,0]
new_y = data_proj[:,1]
plt.scatter(X, Y, color = 'm')
plt.xlabel('x')
plt.ylabel('y')
plt.suptitle('Original dimensions')
plt.show()
plt.scatter(new_x, new_y, color='g' )
plt.xlabel('x_new')
plt.ylabel('y_new')
plt.suptitle('New dimensions')
plt.show()
```

Iris data set with PCA from 4D to 2D.

```python
import numpy as np
import matplotlib.pyplot as plt
from numpy import linalg as LA
from sklearn.datasets import load_iris
iris = load_iris()
data = iris.data 
tar = iris.target
X1 = data[:,0]
X2 = data[:,1]
X2 = data[:,2]
X2 = data[:,3]
mean1 = np.mean(data, axis = 0)
data = data - mean1
cov = (np.dot(np.transpose(data), data)) / (len(data))
w , v = LA.eig(cov)
u = np.array(v[:,[0,1]])
data_proj = np.dot(data, u)
new_x = data_proj[:,0]
new_y = data_proj[:,1]
color = [' '] * len(tar)

for i in range(len(tar)):
    if tar[i] == 0:
        color[i] = 'red' 
    elif tar[i] == 1:
        color[i] = 'green'
    else:
        color[i] = 'yellow'
plt.scatter(new_x, new_y, color=color )
plt.xlabel('x_new')
plt.ylabel('y_new')
plt.suptitle('New dimensions after dimension reduction')
plt.show()
```
