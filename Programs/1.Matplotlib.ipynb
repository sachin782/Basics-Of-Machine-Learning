# 3: MATPLOTLIB

Matplotlib is a numerical mathematics extension Numpy and a great package to view or present data in a pictorial or graphical format. It enables analysts and decision makers to see analytics presented visually, so they can grasp difficult concepts or identify new patterns.

```python
import numpy as np
from matplotlib import pyplot as plt
import pandas as pd
```

Here we are entering student's id as x-axis and test scores as y-axis.Then we are plotting a bar graph.
Note that we need to close the figure using show(),if not closed any follow up plot commands will use same figure.

```python
x = np.arange(5) # assume there are 5 students
y = (20, 35, 30, 35, 27) # their test scores
plt.bar(x,y)
plt.show()
```

Here we are showing a scatter plot with student's id as x-axis and y as y-axis.

```python
plt.scatter(x,y) 
plt.show()
```

Here we are reading iris datasets csv file and using it to plot histograms for petal length, petal width, sepal length and sepal width. These show number of flowers having a specific dimension as per the attribute.Here number of flowers is y-axis and attribute as x-axis.

```python
df = pd.read_csv('iris.csv')
df.hist()# Histogram
plt.show()
```

Here we are using line graph plot with different colour lines for different attributes like green line for petal length.Here number of flowers is y-axis and attribute as x-axis.

```python
df.plot() # Line Graph
plt.show()
```

Box Plot is used to show the shape of the distribution, its central value, and its variability. In a box and whisker plot: the ends of the box are the upper and lower quartiles, so the box spans the interquartile range.The median is marked by a vertical line inside the boxHere we are constructing boxplot for all the attributes in iris data set.The circles outside box are the outlier data sets.

```python
df.boxplot() #Box Plot
plt.show()
```
